```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Store csv into dataframe
disaster_assistance_file = "../Resources/disaster_assistance.csv"
disaster_assistance_df = pd.read_csv(disaster_assistance_file)
disaster_assistance_df.head()
```

```python
disaster_assistance_df = disaster_assistance_df.reset_index().rename({'index':'index'}, axis = 'columns')
disaster_assistance_df = disaster_assistance_df.set_index('index')
disaster_assistance_df.head()
```

```python
#create filtered dataframe from specific columns
disaster_assistance_cols = ["fema_declaration_string", "state", "declaration_date", "incident_type", "declaration_title", "ih_program_declared", "ia_program_declared", "pa_program_declared", "hm_program_declared"]
disaster_assistance_transformed = disaster_assistance_df[disaster_assistance_cols].copy()
disaster_assistance_transformed.head()
```

```python
#rename the column headers
disaster_assistance_transformed = disaster_assistance_transformed.rename(columns={"fema_declaration_string": "fema_declaration",
                                                                                 "ih_program_declared": "indiv_&_household_assistance",
                                                                                 "ia_program_declared": "individual_assistance",
                                                                                 "pa_program_declared": "public_assistance",
                                                                                 "hm_program_declared": "hazard_mitigation"})
disaster_assistance_transformed.head()
```

```python
#store 2nd csv into dataframe
us_disaster_declarations_file = "../Resources/us_disaster_declarations.csv"
us_disaster_declarations_df = pd.read_csv(us_disaster_declarations_file)
us_disaster_declarations_df.head()
```

```python
us_disaster_declarations_df = us_disaster_declarations_df.reset_index().rename({'index':'index'}, axis = 'columns')
us_disaster_declarations_df = us_disaster_declarations_df.set_index('index')
us_disaster_declarations_df.head()
```

```python
#create filtered dataframe from specific columns
us_disaster_declarations_cols = ["fema_declaration_string", "state", "declaration_date", "incident_type", "declaration_title", "incident_begin_date", "incident_end_date", "disaster_closeout_date", "designated_area"]
us_disaster_declarations_transformed = us_disaster_declarations_df[us_disaster_declarations_cols].copy()
us_disaster_declarations_transformed.head()
```

```python
#rename the column headers
us_disaster_declarations_transformed = us_disaster_declarations_transformed.rename(columns={"fema_declaration_string": "fema_declaration",
                                                                                 "incident_begin_date": "begin_date",
                                                                                 "incident_end_date": "end_date",
                                                                                 "disaster_closeout_date": "disaster_closed"})
us_disaster_declarations_transformed.head()
```

```python
#create database connection
connection_string = "{user_name}:{password}@localhost:5432/Natural_Disaster_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#confirm tables
engine.table_names()
```

```python
#load dataframes into database
us_disaster_declarations_transformed.to_sql(name='us_disaster_declarations_transformed', con=engine, if_exists='append', index=True)
```

```python
disaster_assistance_transformed.to_sql(name='disaster_assistance_transformed', con=engine, if_exists='append', index=True)
```

```python
#what are the disaster types and number occurred?
disaster_types = us_disaster_declarations_transformed['incident_type'].unique()
print("Types of Disasters reported:  ", disaster_types)

print("Disaster Occurrences:   ", us_disaster_declarations_transformed['incident_type'].value_counts())
```

```python
#what are the declarations titles for the "other" incidents?
other_types = us_disaster_declarations_transformed.query('incident_type== "Other"')['declaration_title'].value_counts().reset_index()
print(other_types)
```
